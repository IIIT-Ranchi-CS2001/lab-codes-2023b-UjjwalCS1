**Question 1: Visualizing Assembly Election 2023 Results**

Using the following Assembly Election 2023 results, create visualizations to represent the party-wise seat share:

**Election Results:**
**Madhya Pradesh:**
- BJP: Win (163)  
- INC: Win (66)  
- BSP: Win (0)  
- Others: Win (1)

**Rajasthan:**
- INC: Win (69)  
- BJP: Win (115)  
- BSP: Win (2)  
- Others: Win (13)

Tasks:
1. **Two Pie Charts:**  
   - Create two different pie charts (one for Madhya Pradesh and one for Rajasthan) on two different subplots of the same figure object.  
   - Display the percentage seat share on each wedge.  
   - The party with the highest percentage should be shown as slightly detached.

2. **Bar Chart:**  
   - Create a bar chart with party names on the X-axis and seats won on the Y-axis.  
   - Display results for both states on the same bar plot.  
   - Use proper legends to distinguish between the states.


```python
import matplotlib.pyplot as plt

# Data for Assembly Elections 2023
states = ['Madhya Pradesh', 'Rajasthan']
parties = ['BJP', 'INC', 'BSP', 'Others']
mp_seats = [163, 66, 0, 1]
raj_seats = [115, 69, 2, 13]

# Pie chart for Madhya Pradesh
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Madhya Pradesh Pie Chart
ax1.pie(mp_seats, labels=parties, autopct='%1.1f%%', startangle=90, explode=(0.1, 0, 0, 0))
ax1.set_title('Madhya Pradesh Election - 2023')

# Rajasthan Pie Chart
ax2.pie(raj_seats, labels=parties, autopct='%1.1f%%', startangle=90, explode=(0.1, 0, 0, 0))
ax2.set_title('Rajasthan Election - 2023')

# Display Pie charts
plt.show()

# Bar chart
plt.figure(figsize=(10, 6))
bar_width = 0.35
index = range(len(parties))

plt.bar(index, mp_seats, bar_width, label='Madhya Pradesh', color='blue')
plt.bar([i + bar_width for i in index], raj_seats, bar_width, label='Rajasthan', color='red')

plt.xlabel('Parties')
plt.ylabel('Seats Won')
plt.title('Seats Won by Parties in Madhya Pradesh and Rajasthan')
plt.xticks([i + bar_width / 2 for i in index], parties)
plt.legend()
plt.show()
```

**Question 2: Random Number Analysis and Visualization**

Tasks:
1. **Generate a List of Random Numbers:**  
   - Form a list of K random numbers within a limit N where K and N are set by the user. Minimum value of K should be 10.
   - Define a function (or two functions) to return the composite numbers and prime numbers of this list as two separate lists.
   - Determine the square of all prime numbers and square root of all composite numbers
   - Plot both prime numbers Vs their squares and composites Vs their square roots on the same figure object as scatterplots. ( two different subplots on the same figure object)



```python
import random
import math
import matplotlib.pyplot as plt

# Generate K random numbers within limit N
K = int(input("Enter the number of random numbers (minimum 10): "))
N = int(input("Enter the upper limit (N): "))

if K < 10:
    K = 10  # Minimum value of K should be 10

random_numbers = [random.randint(1, N) for _ in range(K)]

# Separate prime and composite numbers
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

prime_numbers = list(filter(is_prime, random_numbers))
composite_numbers = list(filter(lambda x: not is_prime(x), random_numbers))

# Calculate squares for prime numbers and square roots for composite numbers
prime_squares = list(map(lambda x: x ** 2, prime_numbers))
composite_roots = list(map(lambda x: math.sqrt(x), composite_numbers))

# Display the results
print(f"Prime Numbers: {prime_numbers}")
print(f"Squares of Prime Numbers: {prime_squares}")
print(f"Composite Numbers: {composite_numbers}")
print(f"Square Roots of Composite Numbers: {composite_roots}")

# Plotting the results
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot for Prime Numbers vs Squares
ax1.scatter(prime_numbers, prime_squares, color='green')
ax1.set_title('Prime Numbers vs Squares')
ax1.set_xlabel('Prime Numbers')
ax1.set_ylabel('Squares')

# Plot for Composite Numbers vs Square Roots
ax2.scatter(composite_numbers, composite_roots, color='red')
ax2.set_title('Composite Numbers vs Square Roots')
ax2.set_xlabel('Composite Numbers')
ax2.set_ylabel('Square Roots')

plt.show()
```
